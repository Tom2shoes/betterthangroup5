```python
import pandas as pd
import numpy as np
```

```python
# File to Load 
runner_2015 = "Resources/marathon_results_2015.csv"
runner_2016 = "Resources/marathon_results_2016.csv"
runner_2017 = "Resources/marathon_results_2017.csv"
```

```python
# File to Read
runner_2015_df = pd.read_csv(runner_2015)
runner_2016_df = pd.read_csv(runner_2016)
runner_2017_df = pd.read_csv(runner_2017)
```

```python
# Show all 2015 columns
runner_2015_df.columns
```

```python
# Show all 2016 columns
runner_2016_df.columns
```

```python
# Show all 2017 columns
runner_2017_df.columns
```

```python
# Display data frame to see what information it contains
runner_2015_df.head()
```

```python
# Only keep the columns we needed
runner_2015_df = runner_2015_df[["Bib", "Name", "Age", "M/F", "City", "State", "Country", 
                                       "5K", "10K", "15K", "20K", "Half", "25K","30K", "35K", 
                                       "40K", "Pace", "Official Time", "Overall"]]
# Adding Year column
runner_2015_df["Year"] = 2015

# Export csv file
runner_2015_df.to_csv("runner_2015_df.csv")

runner_2015_df.head()
```

```python
# Only keep the columns we needed
runner_2016_df = runner_2016_df[["Bib", "Name", "Age", "M/F", "City", "State", "Country", 
                                       "5K", "10K", "15K", "20K", "Half", "25K","30K", "35K", 
                                       "40K", "Pace", "Official Time", "Overall"]]
# Adding Year column
runner_2016_df["Year"] = 2016

# Export csv file
runner_2016_df.to_csv("runner_2016_df.csv")

runner_2016_df.head()
```

```python
# Only keep the columns we needed
runner_2017_df = runner_2017_df[["Bib", "Name", "Age", "M/F", "City", "State", "Country", 
                                       "5K", "10K", "15K", "20K", "Half", "25K","30K", "35K", 
                                       "40K", "Pace", "Official Time", "Overall"]]
# Adding Year column
runner_2017_df["Year"] = 2017

# Export csv file
runner_2017_df.to_csv("runner_2017_df.csv")

runner_2017_df.head()
```

```python
# Concatenate all DataFrame together
final_df = pd.concat([runner_2015_df, runner_2016_df, runner_2017_df])
```

```python
# Display combined result
final_df
```

```python
# Export final file to csv
final_df.to_csv("Year_Combined.csv")
```
