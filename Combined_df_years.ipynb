```python
import pandas as pd
import numpy as np
```

```python
# File to Load 
runner_2015 = "Resources/marathon_results_2015.csv"
runner_2016 = "Resources/marathon_results_2016.csv"
runner_2017 = "Resources/marathon_results_2017.csv"
```

```python
# File to Read
runner_2015_table = pd.read_csv(runner_2015)
runner_2016_table = pd.read_csv(runner_2016)
runner_2017_table = pd.read_csv(runner_2017)
```

```python
runner_2015_table = runner_2015_table[['Bib', 'Name', 'Age', 'M/F', 'City', 'State', 'Country',
       'Citizen', '5K', '10K', '15K', '20K', 'Half', '25K',
       '30K', '35K', '40K', 'Pace', 'Proj Time', 'Official Time', 'Overall',
       'Gender', 'Division']]
runner_2015_table.head()
```

```python
runner_2017_table = runner_2015_table[['Bib', 'Name', 'Age', 'M/F', 'City', 'State', 'Country',
       'Citizen', '5K', '10K', '15K', '20K', 'Half', '25K',
       '30K', '35K', '40K', 'Pace', 'Proj Time', 'Official Time', 'Overall',
       'Gender', 'Division']]
runner_2017_table.head()
```

```python
final_df = pd.concat([runner_2015_table, runner_2016_table, runner_2017_table], axis=1, join='inner').sort_index()
```
