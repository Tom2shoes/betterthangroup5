```python
import pandas as pd
import numpy as np
```

```python
second = "2015_df.csv"
```

```python
insecond_pd = pd.read_csv(second)
insecond_pd.head()
```

```python
def time_convert(x):
    times = x.split(':')
    return (3600*int(times[0])+60*int(times[1]))+int(times[2])
```

```python
insecond_pd['5k'] = insecond_pd['5K'].apply(time_convert)
insecond_pd['10k'] = insecond_pd['10K'].apply(time_convert)
insecond_pd['15k'] = insecond_pd['15K'].apply(time_convert)
insecond_pd['20k'] = insecond_pd['20K'].apply(time_convert)
insecond_pd['25k'] = insecond_pd['25K'].apply(time_convert)
insecond_pd['30k'] = insecond_pd['30K'].apply(time_convert)
insecond_pd['35k'] = insecond_pd['35K'].apply(time_convert)
insecond_pd['40k'] = insecond_pd['40K'].apply(time_convert)
insecond_pd['Official-Time'] = insecond_pd['Official Time'].apply(time_convert)
```

```python
insecond_pd.head()
```

```python
secondclean = insecond_pd[["Overall","Name","M/F","5k","10k",
                           "15k","20k","25k","30k","35k","40k", "Official-Time"]]
secondclean.head()
```

```python
secondclean.to_csv("2015_in_seconds.csv", index=False, header=True)
```
